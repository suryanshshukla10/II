```python
import matplotlib.pyplot as plt
from skimage import data, io, filters
from skimage import data
from skimage.color import rgb2gray

original = data.astronaut()
grayscale = rgb2gray(original)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

```python
def img_pixel(i):
    x = np.linspace(0,len(grayscale),len(grayscale[i]), dtype=int)
    plt.plot(x,grayscale[i])
    plt.xlabel("Pixel location")
    plt.ylabel("Pixel value")
    
img_pixel(2)
img_pixel(1)
img_pixel(3)
img_pixel(4)
img_pixel(5)
```

```python
import pickle

my_dict = { 1: img1, 2: img2, 3: img3, 4: img4}

def saving_data(fileName='data.json',dataDict=my_dict):
    with open(fileName, 'wb') as fp:
        pickle.dump(dataDict, fp)
        
def read_data(fileName='data.json'):
    with open(fileName, 'rb') as fp:
        data = pickle.load(fp)        
    return data 

saving_data(fileName='part1.json',dataDict=my_dict)
img = read_data('part1.json')
plt.imshow(img[2])
```
